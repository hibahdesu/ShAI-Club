** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
#Creating the fig
fig = plt.figure(figsize=(8, 5))

#Adding ax1
ax1 = fig.add_axes([0,0,1,1])

#Plotting x and z
ax1.plot(x, z)

#Setting the label of x
ax1.set_xlabel('X')

#Setting the label of z
ax1.set_ylabel('Z')

#Adding axs
ax2 = fig.add_axes( [0.2,0.5,.4,.4])

#Plotting x and y
ax2.plot(x, y)

ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)


#Setting the tile of ax2
ax2.set_title('zoom')
#Setting the label of x
ax2.set_xlabel('X')

#Setting the label of z
ax2.set_ylabel('Y')


#Showing the plot
plt.show()
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
#Creating the subplots
fig, axes = plt.subplots(nrows=1, ncols=2)

#Showing them
plt.show()
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
#Creating the subplots
fig, axes = plt.subplots(nrows=1, ncols=2)


#Creating the first subplot
axes[0].plot(x, y, color='b', lw=4, ls='--')
#Creating the seconde subplot
axes[1].plot(x, z, color='r', lw=4, ls='-')

plt.show()
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(13,2))

axes[0].plot(x,y,color="blue", lw=5, ls='-')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
