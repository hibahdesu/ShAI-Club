#### 2. Read all product sales data and display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Note:**It must be implemented in three methods  
The graph should look like this.


```python
df.columns
```

```python
# Create the multiline plots using Matplotlib visualization
plt.plot(df['month_number'], df['facecream'], marker='o', label='Face cream Sales Data', linewidth=3)
plt.plot(df['month_number'], df['facewash'], marker='o', label='Face Wash Sales Data',linewidth=3)
plt.plot(df['month_number'], df['toothpaste'], marker='o', label='ToothPaste Sales Data', linewidth=3)
plt.plot(df['month_number'], df['bathingsoap'], marker='o', label='ToothPaste Sales Data', linewidth=3)
plt.plot(df['month_number'], df['shampoo'], marker='o', label='ToothPaste Sales Data', linewidth=3)
plt.plot(df['month_number'], df['moisturizer'], marker='o', label='ToothPaste Sales Data', linewidth=3)

#Chaning the y axes
plt.yticks([1000, 2000, 4000, 6000, 8000,10000, 12000, 15000, 18000])

#Creating the legend
plt.legend(loc='upper left')

#Setting the title
plt.title('Sales data')

#Setting the xlabel
plt.xlabel('Month Number')

#Setting the ylabel
plt.ylabel('Sales units in number')

plt.xticks(df['month_number'])

# Showing the plot
plt.show()
```

```python
############################## multiline plots using Matplotlib visualization #########################################
```

```python
# Create the multiline plots Plot using Seaborn visualization
sns.lineplot(df['month_number'], df['facecream'], marker='o', label='Face cream Sales Data', linewidth=3)
sns.lineplot(df['month_number'], df['facewash'], marker='o', label='Face Wash Sales Data', linewidth=3)
sns.lineplot(df['month_number'], df['toothpaste'], marker='o', label='ToothPaste Sales Data', linewidth=3)
sns.lineplot(df['month_number'], df['bathingsoap'], marker='o', label='ToothPaste Sales Data', linewidth=3)
sns.lineplot(df['month_number'], df['shampoo'], marker='o', label='ToothPaste Sales Data', linewidth=3)
sns.lineplot(df['month_number'], df['moisturizer'], marker='o', label='ToothPaste Sales Data', linewidth=3)


#Chaning the y axes
plt.yticks([1000, 2000, 4000, 6000, 8000,10000, 12000, 15000, 18000])


#Creating the legend
plt.legend(loc='upper left')

#Setting the title
plt.title('Sales data')

#Setting the xlabel
plt.xlabel('Month Number')

#Setting the ylabel
plt.ylabel('Sales units in number')

plt.xticks(df['month_number'])

# Showing the plot
plt.show()
```

```python
########################### multiline plots Plot using Seaborn visualization #########################################
```

```python
# Create the multiline plot using Pandas visualization
df.plot(x='month_number', y=['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer'], marker='o', linewidth=3)


plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])

# creating the lengend
plt.legend(loc='upper left')

#Setting the title
plt.title('Sales data')

#Setting the xlabel
plt.xlabel('Month Number')

#Setting the ylabel
plt.ylabel('Sales units in number')

plt.xticks(df['month_number'])

# Showing the plot
plt.show()
```

    - In Pie chart display Number of units sold per year for each product in percentage.
    - It must be implemented in two methods.

The Pie chart should look like this.


```python
df.columns
```

```python
df['facecream'].sum()
df['facewash'].sum()
df['toothpaste'].sum()
df['bathingsoap'].sum()
df['shampoo'].sum()
df['moisturizer'].sum()
```

```python
#Creating the data I am going to put
columns = [df['facecream'].sum(),
df['facewash'].sum(),
df['toothpaste'].sum(),
df['bathingsoap'].sum(),
df['shampoo'].sum(),
df['moisturizer'].sum()]

plt.pie(columns, labels=['Facecream', 'Facewash', 'Toothpaste', 'Bathingsoap',
       'Shampoo', 'Moisturizer'], autopct='%1.1f%%')

#Putting the title
plt.title('Sales data')

#Putting the legend
plt.legend(loc='lower right')

#Showing the plot
plt.show()
```
